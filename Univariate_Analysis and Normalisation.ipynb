```python
q1 = df['ApplicantIncome'].quantile(.25)
print('q1:',q1)
q3 = df['ApplicantIncome'].quantile(.75)
print('q3:',q3)

#Inter-quartile range
IQR = q3-q1
print('IQR:',IQR)
```

```python
q1 = np.log(df['ApplicantIncome'].quantile(.25))
print('q1:',q1)
q3 = np.log(df['ApplicantIncome'].quantile(.75))
print('q3:',q3)

#Inter-quartile range
IQR = q3-q1
print('IQR:',IQR)
```

```python
#calculating standard deviation
import statistics as st

sd = st.stdev(np.log(df['ApplicantIncome']))
sd
```

```python
#calculating Sigma
#this should be 1.35 to be normalized
sigma=IQR/sd
print(sigma)
```
